```python
 from PIL import Image
 import numpy as np
```

```python
# Question 1
img  = Image.open("surfacePic.jpg")  
```

```python
pix = np.array(img)
```

```python
#Question 2
pix.shape
```

```python
print(pix)
```

```python
#Question 2.a
import matplotlib.pyplot as pyplot
pyplot.imshow(pix)
pyplot.show()
```

```python
#Question 2.b
R=np.copy(pix)
R[:,:,1:]=0
```

```python
G=np.copy(pix)
G[:,:,[0,2]]=0
```

```python
B=np.copy(pix)
B[:,:,[0,1]]=0
```

```python
pyplot.subplot(1,3,1, title="Red Channel")
pyplot.imshow(R)
pyplot.subplot(1,3,2, title="Green Channel")
pyplot.imshow(G)
pyplot.subplot(1,3,3, title="Blue Channel")
pyplot.imshow(B)
pyplot.tight_layout()

pyplot.show()
```

```python
#Question 3
R_avg = R[:,:,0].mean(0)
G_avg = G[:,:,1].mean(0)
B_avg = B[:,:,2].mean(0)
```

```python
pyplot.plot(range(0,512), R_avg, color="red")
pyplot.plot(range(0,512), G_avg, color="green")
pyplot.plot(range(0,512), B_avg, color="blue")
pyplot.title("Column Averages of Each Channel")
pyplot.show()
```

```python
#Question 4 : Above half - below half
R_diff1 = R[0:256]-R[256:512]
R_diff2 = R[256:512]-R[0:256]
R_diff = np.concatenate([R_diff1,R_diff2])
G_diff1 = G[0:256]-G[256:512]
G_diff2 = G[256:512]-G[0:256]
G_diff = np.concatenate([G_diff1,G_diff2])
B_diff1 = B[0:256]-B[256:512]
B_diff2 = B[256:512]-B[0:256]
B_diff = np.concatenate([B_diff1,B_diff2])
```

```python
pyplot.subplot(1,3,1, title="Red Channel")
pyplot.imshow(R_diff)
pyplot.subplot(1,3,2, title="Green Channel")
pyplot.imshow(G_diff)
pyplot.subplot(1,3,3, title="Blue Channel")
pyplot.imshow(B_diff)
pyplot.tight_layout()

pyplot.show()
```
